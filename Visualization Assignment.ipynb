```python
import pandas as pd
%matplotlib inline
df1 = pd.read_csv('E:/Competing through BA/Visualization/ConstructionTimeSeriesDataV2.csv')
df1.head()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(df1.Month, df1["Total Construction"], label='Total Construction')
ax.plot(df1.Month, df1["Private Construction"], label="Pirvate Construction")
ax.plot(df1.Month, df1["Public Construction"], label="Public Construction")
plt.xlabel("Month")
plt.ylabel("Construction Spending")
ax.legend(loc='upper right', prop={'family':'Times New Roman','size':'large'})
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig.figsize=(30,20)
fig.suptitle('Construction Spending Overall',fontsize='xx-large')
plt.savefig('ConstructionOverall.jpg', bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x_labels=['','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
x=df1.Month[0:12]
y1=df1["Total Construction"][0:12]
y2=df1["Private Construction"][0:12]
y3= df1["Public Construction"][0:12]
ax.plot(x, y1, label='Total Construction')
ax.plot(x, y3, label="Public Construction")
plt.ylabel("Construction Spending")
ax.xaxis.set_ticks(range(0,13))
ax.xaxis.set_ticklabels(x_labels)
ax.legend(loc='upper right', prop={'family':'Times New Roman','size':'large'})
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlim(0,x.max()+1)
fig.suptitle('Construction Spending the 1st Year',fontsize='xx-large')
plt.savefig('ConstructionSpending1stYear.jpg')
```
